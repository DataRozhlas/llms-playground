```python
local_llms = {}
for m in models:
    local_llms[m['model']] = [m['details']['parameter_size'], m['details']['quantization_level'], m['modified_at']]
```

## ƒå√≠m to testujeme 

```python
system_prompts = ['Jsi inteligentn√≠, v≈°estrannƒõ vzdƒõlan√°, seƒçtƒõl√° a bystr√° asistentka. Bude≈° odpov√≠dat na moje dotazy, spisovnou ƒçe≈°tinou, zdvo≈ôile, srozumitelnƒõ a jasnƒõ.']
```

```python
otazky = [
    
    'Jednou vƒõtou vysvƒõtli n√°sleduj√≠c√≠ pojmy: 1/ N√°dra≈æka. 2/ Od√∫mr≈•. 3/ Pacht. 4/ Pacholek. 5/ Prostop√°≈°n√≠k. 6/ Hovado. 7/ ≈Ω√°kajda. 8/ Turnaj stolet√≠. 9/ Pas≈•√°k. 10/ Kofola.',
    'Uveƒè jm√©na pƒõti slavn√Ωch ƒçesk√Ωch spisovatel≈Ø ƒçi spisovatelek.',
    'Kam se v ƒåesku jezd√≠ ly≈æovat?',
    'O ka≈æd√© z tƒõchto ≈æen jednou vƒõtou ≈ôekni, ƒç√≠m se proslavila: Milada Hor√°kov√°, Marie Kude≈ô√≠kov√°, Vƒõra Chytilov√°.',
    'Kdo je posledn√≠m (nejnovƒõj≈°√≠m) tobƒõ zn√°m√Ωm zn√°m√Ωm p≈ôedsedou ƒçesk√© vl√°dy, tedy premi√©rem ƒåesk√© republiky?',
    'V≈Øƒçi kter√Ωm n√°rodnostn√≠m nebo etnick√Ωm men≈°in√°m projevuje p≈ôedsudky nejvƒõt≈°√≠ ƒç√°st vƒõt≈°inov√© populace v ƒåesku?',
    'Uveƒè t≈ôi dlouhodobƒõ √∫spƒõ≈°n√© ƒçesk√© technologick√© (internetov√©) firmy.',
    'Jak√° mal√° j√≠dla se v ƒçesk√Ωch hospod√°ch, a to i tƒõch bez tepl√© kuchynƒõ, pod√°vaj√≠ k pivu?'
    
]
```

```python
len(otazky)
```

```python
odpovedi = []
for o in otazky:
    print(o.upper())
    for m, val in local_llms.items():
        if m != 'gemma:latest':
            for s in system_prompts:
                print(f"\n{m.upper()} | system prompt: {s}\n")
                odpoved = {}
                if s != None:
                    resp = ollama.generate(model=m, prompt=o, system=s)
                else:
                    resp = ollama.generate(model=m, prompt=o)
                odpoved['model'] = m
                odpoved['systemovy_prompt'] = s
                odpoved['prompt'] = o
                odpoved['odpoved'] = resp['response']
                odpoved['cas'] = resp['total_duration']
                odpoved['params'] = val[0]
                odpoved['quant'] = val[1]
                odpoved['verze'] = val[2]
                print(odpoved['odpoved'])
                odpovedi.append(odpoved)
```

```python
vysledky = pd.DataFrame(odpovedi)
```
