Rychlý průzkum schopností LLMs před prezentací v PAQ Research. Otázky míří na všeobecné znalosti o Česku a schopnost komunikovat v češtině.

```python
import time
```

```python
import ollama
```

```python
local_llms = [
    'mistral', 'llama2', 'llama2:13b', 'mistral-openorca', 'vicuna', 'solar'
]
```

Modely zkoušíme defaultní, bez extra kvantizace atd.

```python
systemovy_prompt = 'Jsi redaktorka se skvělým všeobecným přehledem a vynikající znalostí českého jazyka. Na dotazy a úkoly reaguješ spisovnou češtinou, zdvořile a co nejstručněji.'
```

Od počátku komunikujeme s LLMs česky a nutíme je do českých odpovědí.

```python
otazky = [
    
    'Uveď jména pěti slavných českých spisovatelů či spisovatelek.',
    'Jaká řeka protéka pod Karlovým mostem?',
    'Která ulice v Ostravě je vyhlášená množstvím podniků a bujarým nočním životem?',
    'Čím byl v československých dějinách významný Edvard Beneš? Odpověz nejvýše třemi větami.',
    'Kdo je posledním (nejnovějším) tobě známým známým předsedou české vlády, tedy premiérem České republiky?',
    'Vůči kterým národnostním nebo etnickým menšinám projevuje předsudky největší část většinové populace v Česku?',
    'Uveď tři dlouhodobě úspěšné české technologické (internetové) firmy.',
    'Jaké jsou dlouhodobé problémy české ekonomiky? Jaká politická opatření by je pomohla zmenšit či překonat?'

]
```

Níže uváděné časy měřím na notebooku s i5 Alder Lake a 16 GB RAM.

```python
for o in otazky:
    print(f"{o.upper()}\n")
    for m in local_llms:
        modelfile=f"""
FROM {m}
SYSTEM {systemovy_prompt}
"""
        ollama.create(model='example', modelfile=modelfile)

        start_time = time.time()
        response = ollama.chat(model='example', messages=[{'role': 'user','content': f'''{o}''',},])
        stopky = time.time() - start_time
        print(f"{m.upper()} ({round(stopky,2)} s)\n")
        print(f"{response['message']['content']}\n")
```
