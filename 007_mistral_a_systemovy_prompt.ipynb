```python
import ollama
import pandas as pd
```

```python
models = ollama.list()['models'][1:]
models
```

```python
local_llms = {}
for m in models:
    if "mistral:latest" in m['model']:
        local_llms[m['model']] = [m['details']['parameter_size'], m['details']['quantization_level'], m['modified_at']]
local_llms
```

```python
system_prompts = [
    'Jsi úslužná asistentka.',
    'Jsi úslužná asistentka. Komunikuješ spisovnou češtinou.',
    'Jsi vzdělaná literární vědkyně s doktorátem z Karlovy univerzity.',
    'Jsi super inteligentní umělá inteligence. Odpovídáš po pravdě, na základě nejlepších zdrojů.',
    'Jsi super inteligentní umělá inteligence. Odpovídáš po pravdě, na základě nejlepších zdrojů. Komunikuješ v českém jazyce, spisovně, slušně a především fakticky správně.',
    'Skládáš maturitní zkoušku z českého jazyka. Následuje otázka, na kterou co nejlépe odpovíš.',
    'Skládáš maturitní zkoušku z českého jazyka. Na tom, jak správně odpovíš, závisí kvalita celého tvého života. Když odpovíš správně, půjdeš na vysokou školu a budeš mít hezký život. Když odpovíš špatně, zemřeš pod mostem.',
    'Skládáš maturitní zkoušku z českého jazyka. Na tom, jak správně odpovíš, závisí kvalita celého tvého života. Když odpovíš správně, půjdeš na vysokou školu a budeš mít hezký život. Když odpovíš špatně, zemřeš pod mostem. Jde o zkoušku z českého jazyka, vedle faktické správnosti odpovědí záleží i na tom, jak spisovnou a bezchybnou češtinou na ni odpovíš.',
    'Jsi knihovnice. Máš přístup do skvěle zásobené knihovny, najdeš v ní spoustu informací. S pomocí kvalitní literatury mi odpověz na následující otázku.'
                 ]
system_prompts
```

```python
otazky = [
    
    f"""Uveď jména tří významných českých spisovatelů a tří významných českých spisovatelek."""

]
```

```python
odpovedi = []
for o in otazky:
    print(o.upper())
    for m, val in local_llms.items():
        for s in system_prompts:
            print(f"\n{m.upper()} | system prompt: {s}\n")
            odpoved = {}
            if s != None:
                resp = ollama.generate(model=m, prompt=o, system=s)
            else:
                resp = ollama.generate(model=m, prompt=o)
            odpoved['model'] = m
            odpoved['systemovy_prompt'] = s
            odpoved['prompt'] = o
            odpoved['odpoved'] = resp['response']
            odpoved['cas'] = resp['total_duration']
            odpoved['params'] = val[0]
            odpoved['quant'] = val[1]
            odpoved['verze'] = val[2]
            print(odpoved['odpoved'])
            odpovedi.append(odpoved)
```

```python
frejm = pd.DataFrame(odpovedi)
frejm
```
