```python
import ollama
import pandas as pd
models = ollama.list()['models'][1:]
models
local_llms = {}
for m in models:
    local_llms[m['model']] = [m['details']['parameter_size'], m['details']['quantization_level'], m['modified_at']]
local_llms
```

```python
system_prompts = [
    "You are going to act as a skilled professional translator. I will provide you with a text in a foreign language, and you will reply with a translation into English."
                 ]
```

```python
otazky = [
    
    f"""Z žádných historických dokumentů či zpráv starověkých autorů nelze vyčíst, že by někdy jakákoli společnost před vznikem peněz fungovala na základě směnných obchodů, jako je „já ti dám slepice, ty mi dej krávu“. Antropologové ani neobjevili žádné domorodce – ať už africké, jiho- či severoamerické nebo australské –, kteří by podobným způsobem uspokojovali své každodenní potřeby."""

]
```

```python
odpovedi = []
for o in otazky:
    print(o.upper())
    for m, val in local_llms.items():
        if m != 'gemma:latest':
            for s in system_prompts:
                print(f"\n{m.upper()} | system prompt: {s}\n")
                odpoved = {}
                if s != None:
                    resp = ollama.generate(model=m, prompt=o, system=s)
                else:
                    resp = ollama.generate(model=m, prompt=o)
                odpoved['model'] = m
                odpoved['systemovy_prompt'] = s
                odpoved['prompt'] = o
                odpoved['odpoved'] = resp['response']
                odpoved['cas'] = resp['total_duration']
                odpoved['params'] = val[0]
                odpoved['quant'] = val[1]
                odpoved['verze'] = val[2]
                print(odpoved['odpoved'])
                odpovedi.append(odpoved)
```

```python
frejm = pd.DataFrame(odpovedi)
frejm
```
